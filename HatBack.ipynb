# Hat Back - Monte Carlo
In class, a very interesting thing was said. Imagine a party where at the begining everyone threw their hats in a room. At the end of the party the power goes out and everyone has to blindly grab a hat and go home. How may people will get their own hat back? Does it change with how many people are there?

The expected value is 1, no matter how many people are there

```python
import json
import numpy as np
import random as ra
import pandas as pd
```

I made a text file with the top 200 baby names

```python
f_names = './data/baby_names.txt'
with open(f_names, 'r') as f:
    names = f.read().splitlines()
```

```python
def Hat_Back(names, iterations):
    """
    Takes:
        names      - list of names
        iterations - how many parties
    Returns:
        The number of people that got their hat back after all iterations
    """
    o_names = names.copy()
    after_all_parties = 0
    for n in range(iterations):
        ra.shuffle(names)
        hat_back = 0
        for i in range(len(names)):
            if names[i] == o_names[i]:
                hat_back += 1
        after_all_parties += hat_back
    return after_all_parties
```

```python
for i in range(2, 200, 1):
    z = Hat_Back(names[:i], 10000)
    # z should be around the number of iterations no matter how many partiers there are
    print("After 10000 parties, with {} people, {} got their hat back.".format(i, z))
```
